```python
#Check for missing values
df.isnull().sum()
```

```python
#Handle Missing Values
for col in df.columns:
    if df[col].dtype in ['float64', 'int64']:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna(df[col].mode()[0])
```

```python
#Standardize Units
area_cols = [c for c in df.columns if 'AREA' in c.upper()]
production_cols = [c for c in df.columns if 'PRODUCTION' in c.upper()]

for col in area_cols:
    df[col] = df[col] * 1000  # Convert thousand hectares ‚Üí hectares

for col in production_cols:
    df[col] = df[col] * 1000  # Convert thousand tons ‚Üí tons
```

```python
#Remove duplicates
df = df.drop_duplicates()
```

```python
#Standardize Text Columns
# Convert State Name and District Name to title case
if 'State Name' in df.columns:
    df['State Name'] = df['State Name'].str.title()

if 'Dist Name' in df.columns:
    df['Dist Name'] = df['Dist Name'].str.title()
```

```python
#Consistency Checks
# Ensure Year is numeric and within valid range
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
df = df[df['Year'].between(1960, 2025)]
```

```python
df.to_csv("Cleaned_Agriculture_Data.csv", index=False)
print("\n‚úÖ Data cleaning completed successfully!")
print("üìÅ Cleaned data saved as: Cleaned_Agriculture_Data.csv")
```
